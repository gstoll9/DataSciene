# Exploratory Data Analysis
***

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import myEDA

data_dir = "./data/"
```

## Data

```python
data = pd.read_csv(data_dir + "isotopes_clean.csv")
print(data.shape)
data.head()
```

```python
data.info()
```

```python
data.describe().transpose()
```

```python
cat_cols = myEDA.getCategoricals(data)
cat_cols.remove("abundance")
cat_cols = np.array(cat_cols)
cat_cols
```

# Distributions
***

## Histograms

```python
# drop categories
df = data.copy()
df["half_life"] = np.log(df["half_life"])
df["abundance"] = df["abundance"].replace(0, np.nan)

myEDA.histograms(df, cat_cols)
```

```python
df["half_life"].hist(bins=50);
```

```python
cats = df["protons"].value_counts().sort_index()
plt.bar(cats.index, cats, width=0.1, color="orange")
plt.plot(cats)
```

## Box Plots

```python
# drop categories
df = data.drop(cat_cols, axis=1)

# log half life
df["half_life"] = np.log(df["half_life"])

# replace 0 with nan
df["abundance"] = df["abundance"].replace(0, np.nan)

# remove outliers
df = df[df["magnetic_moment"] < 20]
df = df[df["quadrupole_moment"] < 13]

myEDA.boxplots(df)
```

```python
df["quadrupole_moment"].plot.box(sym='k_');
```

# Correlation
***

## Scatter Plots

```python
# drop categories
cols = cat_cols[np.where((cat_cols != "protons") & (cat_cols != "neutrons"))]
df = data.drop(cols, axis=1)
df["half_life"] = np.log(df["half_life"])
df["abundance"] = df["abundance"].replace(0, np.nan)
df["binding_energy"] = np.sqrt(df["mass_excess"])
df["magnetic_moment"].replace(0,np.nan)

myEDA.scatterplots(df, color="half_life")
```

```python
df.plot.scatter(x="protons", y="quadrupole_moment", s=1,
                c=np.log(df["half_life"]), cmap="plasma");
```

## Correlation Matrix

```python
# drop categories
df = data.drop(cat_cols[[0,1,2,3]], axis=1)

print(df.columns)

plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), square=True)
plt.savefig("./plots/corrmat.png")
```

# Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

include = ["protons","neutrons",]
df = data[include]

vif = pd.DataFrame({
    "VIF Factor":[variance_inflation_factor(df.values, i) for i in range(df.shape[1])],
    "features":df.columns
})
vif
```
